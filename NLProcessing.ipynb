```python
import pandas as pd
```

```python
lyrics_df = pd.read_csv('billboard_lyrics_1964-2015.csv', encoding='latin-1')
```

```python
lyrics_df.query('Year == 2007')
```

```python
lyrics_df.dtypes
```

```python
aliciaKeys = lyrics_df.loc[lyrics_df['Artist']== 'pink', 'Lyrics'].values[0]
```

```python
print(aliciaKeys)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords 
stopWords = set(stopwords.words("english"))
print(stopWords)
aliciaKeysToken = word_tokenize(aliciaKeys)
```

```python
wordsFiltered=[]
for w in aliciaKeysToken:
    if w not in stopWords:
        wordsFiltered.append(w)
        
wordsFiltered = ' '.join(wordsFiltered)
print(wordsFiltered)
```

```python
import nltk
from nltk import FreqDist
tokensDist = nltk.word_tokenize(wordsFiltered)
freqDistOne = FreqDist(tokensDist)
print(freqDistOne)
print(freqDistOne.most_common(15))
```
